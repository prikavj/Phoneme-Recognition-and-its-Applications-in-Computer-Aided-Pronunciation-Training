# Import libraries

```python
import json
import matplotlib.pyplot as plt
```

# Helper functions

```python
def plot_accuracy_with_epochs(epochs, accuracy, val_accuracy, model_name):
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, accuracy, label='Training Accuracy')
    plt.plot(epochs, val_accuracy, label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training and Validation Accuracy for ' + model_name)
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_loss_with_epochs(epochs, loss, val_loss, model_name):
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, loss, label='Training Loss')
    plt.plot(epochs, val_loss, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss for ' + model_name)
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_per_with_epochs(per_data, model_name):
    # Extract the index and PER values from the data
    index = list(per_data.keys())
    per_values = list(per_data.values())

    # Plot Phoneme Error Rate with respect to the index
    plt.figure(figsize=(8, 6))
    plt.plot(index, per_values, marker='o', linestyle='-', color='b')
    plt.xlabel('Epochs')
    plt.ylabel('Phoneme Error Rate')
    plt.title('Phoneme Error Rate with respect to Epochs for ' + model_name)
    plt.grid(True)
    plt.show()

def plot_metrics_for_model(history_file, per_file, model_name):
    # Load history data from JSON file
    with open(history_file, 'r') as json_file:
        history_data = json.load(json_file)

    # Extract accuracy and loss data from history data
    epochs = history_data["epochs"]
    accuracy = history_data["accuracy"]
    loss = history_data["loss"]
    val_accuracy = history_data["val_acc"]
    val_loss = history_data["val_loss"]

    # Load PER data from JSON file
    with open(per_file, 'r') as json_file:
        per_data = json.load(json_file)

    # Plot accuracy and loss
    print(f"Model: {model_name}")
    plot_accuracy_with_epochs(epochs, accuracy, val_accuracy, model_name)
    plot_loss_with_epochs(epochs, loss, val_loss, model_name)

    # Plot PER
    plot_per_with_epochs(per_data, model_name)

def get_performance_data(filename):
    with open(filename, 'r') as json_file:
          data = json.load(json_file)

    # Extract data
    epochs = data["epochs"]
    accuracy = data["accuracy"]
    loss = data["loss"]
    val_accuracy = data["val_acc"]
    val_loss = data["val_loss"]

    return epochs, accuracy, loss, val_accuracy, val_loss

```

```python
history_filenames = ['history_cnn_rnn.json', 'history_rnn.json', 'history_dnn.json']
per_filenames = ['per_cnn_rnn.json', 'per_rnn.json', 'per_dnn.json']
model_names = ['CNN-RNN', 'RNN', 'DNN']

base_history_files_path = '/content/drive/MyDrive/TextAnalyticsProject/PerformanceAnalysis/history_files/'
base_per_files_path = '/content/drive/MyDrive/TextAnalyticsProject/PerformanceAnalysis/per_files/'
# Creating full history file paths
history_file_paths = [base_history_files_path + filename for filename in history_filenames]
per_file_paths = [base_per_files_path + filename for filename in per_filenames]
```

## Analysis for CNN-RNN

```python
history_file_cnn_rnn = history_file_paths[0]
per_file_cnn_rnn = per_file_paths[0]

# Load history data from JSON file
with open(history_file_cnn_rnn, 'r') as json_file:
    history_data_cnn_rnn = json.load(json_file)

# Load PER data from JSON file
with open(per_file_cnn_rnn, 'r') as json_file:
    per_data_cnn_rnn = json.load(json_file)

model_name = model_names[0]
# Extract accuracy and loss data from history data
epochs = history_data_cnn_rnn["epochs"]
accuracy = history_data_cnn_rnn["accuracy"]
loss = history_data_cnn_rnn["loss"]
val_accuracy = history_data_cnn_rnn["val_acc"]
val_loss = history_data_cnn_rnn["val_loss"]
```

### Accuracy:

```python
# Plot accuracy
print(f"Model: {model_name}")
plot_accuracy_with_epochs(epochs, accuracy, val_accuracy, model_name)
```

### Loss:

```python
# Plot loss
print(f"Model: {model_name}")
plot_loss_with_epochs(epochs, loss, val_loss, model_name)
```

### Phoneme Error Rate:

```python
# Plot PER
print(f"Model: {model_name}")
plot_per_with_epochs(per_data_cnn_rnn, model_name)
```

**Inference**
- The model has been trained for 10 epochs.
- The training accuracy starts at around 38.7% and steadily increases with each epoch, reaching approximately 54.5% by the 10th epoch.
- The training loss decreases over epochs, starting at 2.27 and converging to 1.51 by the 10th epoch.
- The validation accuracy starts at 43.8% and fluctuates throughout training, reaching around 51.6% by the 10th epoch.
- The validation loss starts at 1.99 and slightly decreases, reaching around 1.63 by the 10th epoch.
- The model might be showing signs of overfitting, as the training accuracy is significantly higher than the validation accuracy by the end.
